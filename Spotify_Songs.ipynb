```python
mean_popularity = df['track_popularity'].mean()

df['popularity_results'] = np.where(df['track_popularity'] > mean_popularity, 'P', 'F')

def pie(playlist_genre):
    """This function creates a pie chart with percentages of successful songs in different categories.

    Args:
        category: str. Category name in the Spotify dataset.

    Returns:
        None
    """
    label = []
    label_percent = []

    for cat in df[playlist_genre].unique():
        label.append(cat)
        pass_cat = df[(df[playlist_genre] == cat) & (df['popularity_results'] == 'P')].shape[0]
        all_cat = df[df[playlist_genre] == cat].shape[0]
        label_percent.append(pass_cat / all_cat * 100)

    fig, axs = plt.subplots()
    axs.pie(label_percent, labels=label, autopct='%1.1f%%')
    plt.title(f"Popularity based on {playlist_genre}")
    plt.show()

```
